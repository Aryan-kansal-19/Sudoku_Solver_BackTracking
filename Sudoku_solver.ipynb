```python
board = [
    [7, 8, 0 , 4, 0, 0 , 1, 2, 0],
    [6, 0, 0 , 0, 7, 5 , 0, 0, 9],
    [0, 0, 0 , 6, 0, 1 , 0, 7, 8],
    
    [0, 0, 7 , 0, 4, 0 , 2, 6, 0],
    [0, 0, 1 , 0, 5, 0 , 9, 3, 0],
    [9, 0, 4 , 0, 6, 0 , 0, 0, 5],
    
    [0, 7, 0 , 3, 0, 0 , 0, 1, 2],
    [1, 2, 0 , 0, 0, 7 , 4, 0, 0],
    [0, 4, 9 , 2, 0, 6 , 0, 0, 7]
]
```

```python
def print_board(bo):
    for i in range(len(bo)):
        if i % 3 == 0 and i != 0:
            print("----------------------------------")
        for j in range(len(bo[i])):
            if j % 3 == 0 and j != 0  :
                print(" | ",end = " ")
            if j == 8 :
                    print(bo[i][j])
            else :
                    print(str(bo[i][j])+ " ",end=" ")
    print()
print_board(board)
```

```python
def valid(bo, num, pos):
    # check row
    for r in range(len(bo[0])):
        if bo[pos[0]][r] == num and pos[1] != r:
            return False
        
    # check col
    for c in range(len(bo[0])):
        if bo[c][pos[1]] == num and pos[0] != c:
            return False
    
    # Check Cubes 
    cube_x = pos[1]//3
    cube_y = pos[0]//3
    
    for i in range (cube_y *3, cube_x*3 +3):
        for j in range (cube_x *3, cube_y*3 +3):
            if bo[i][j] == num and (i, j) != pos:
                return False
            
    return True
```

```python
def find_empty(bo):
    for i in range(len(bo)):
        for j in range(len(bo[0])):
            if bo[i][j] == 0:
                return ((i, j))
    return None
```

```python
def solve(bo,c):
    print("Counter = ", c)
    print_board(bo)
    c +=1

    print()
    find = find_empty(board)
    if not find :
        return True
    else:
        row, col = find
    
    for i in range(1,10):
        if valid(board, i, (row, col)):
            bo[row][col] = i
            
            if solve(bo,c):
                return True
            bo[row][col] = 0
    return False
    
```

```python
print_board(board)
a = board
solve(board,0)
print()
print("Final")
print_board(a)
```
